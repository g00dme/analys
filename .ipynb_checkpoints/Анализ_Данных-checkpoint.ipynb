```python
dum=pd.concat((dum,pd.get_dummies(car[['Марка авто','Штат или провинция','Топливо','Тип продавца','Трансмиссия','Владелец']])),axis=1)
dum = dum.dropna()
X = dum.drop('Цена продажи', axis = 1)
Y = dum['Цена продажи']
model = LinearRegression()
model.fit(X, Y)
print('R^2: {0}'.format(model.score(X, Y)))
```

```python
def regression_coef(model, X, Y):
  coef = pd.DataFrame(
      zip(['intercept'] + X.columns.tolist(),
       [model.intercept_] + model.coef_.tolist()),
    columns=['predictor', 'coef'])
X1= np.append(np.ones((len(X),1)), X, axis=1)
b = np.append(model.intercept_, model.coef_)
MSE = np.sum((model.predict(X) - Y) ** 2,
             axis=0) / float(X.shape[0] - X.shape[1])
var_b = MSE * (np.linalg.inv(np.dot(X1.T, X1)).diagonal())
sd_b = np.sqrt(var_b)
t = b / sd_b
coef['pvalue'] = [2 * (1 - stats.t.cdf(np.abs(i),
       (len(X1) - 1))) for i in t]
return coef

regression_coef(model, X, Y)
```
